```python
# 3 Defining Imports

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn import linear_model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression

%matplotlib inline
```

```python
# 4 & 5 Reading data from CSV file and storing it into df_zoo variable

df_zoo = pd.read_csv("zoo.csv")
df_zoo.info()
df_zoo
```

```python
# 6 Reading data from CSV file and storing it into df_zoo variable

df_class = pd.read_csv("class.csv")
df_class.info()
df_class
```

```python
# 7 Merging the data frames

df_zoo = df_zoo.join(df_class.set_index("Class_Number"),on = "class_type")
df_zoo
```

```python
# 8 Droping the unnecessary columns

df_zoo.drop(["catsize","Number_Of_Animal_Species_In_Class","Animal_Names"], axis = 1, inplace = True)
```

```python
# 9 Splitting the data into training and testing data.

x = df_zoo.drop(["animal_name","class_type","Class_Type"],axis=1)
y = df_zoo["class_type"]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
# 10 Creating the modele in order to fit it to the training data.

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(x_train,y_train)
```

```python
# 11 Predicting values based on testing data.

pred = knn.predict(x_test)
```

```python
# 12 Print out the classification report y test data and predictions,

print(classification_report(y_test,pred))
print("Training Score: ",knn.score(x_train,y_train))
print("Testing Score : ",knn.score(x_test,y_test))
```

Observation(13): After running it a couple of times, it seems that when we are passing the value K = 1 the data accuracy is always high and it is lying in between 80% (.80) to 100% (1.00). But it mostly depends on the random picked data and We have noticed that when the precision for all scenarios is equal to 1.00 then accuracy is also 1.00 or 100%.

```python
# 14 Repeating steps 10 to 13 using K = 3

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train,y_train)
prediction=knn.predict(x_test)
print(classification_report(y_test,prediction))
print("Training Score: ",knn.score(x_train,y_train))
print("Testing Score : ",knn.score(x_test,y_test))
```

Observation(14): After running it a couple of times, it seems that when we are passing the value K = 3 the data accuracy is decreasing but it is not getting closer to 100%(1.00). In this case, precision is not equal to 1.00 for all the scenarios, so the accuracy has decreased.

```python
# 15 Creating LogisticRegression model, Fit it to the training data and predicting based on the test data.

logisticRegr = LogisticRegression().fit(x_train,y_train)
predictions=logisticRegr.predict(x_test)
print(classification_report(y_test,predictions))
print("Training Score: ",logisticRegr.score(x_train,y_train))
print("Testing Score : ",logisticRegr.score(x_test,y_test))
```

Observation(16): After running it couple of time, From these three model, it seems that the accuracy of all three Machine learning models depended on the random picked data and if the precision for all the scenarios are equal to 1.00 then the accuracy is always 1.00 or 100%

Observation(17): KNN and Logistic regression are the Machine Learning Models used for Classification. KNN is a distance based technique while Logistic regression is probability based. The k-nearest-neighbours algorithm is a classification algorithm. KNN plots the data points (training data) into a vector space and while predicting it plots your test data-point and finds the k-nearest neighbours. However, logistic regression is a discriminative algorithm, meaning it tries to find boundaries between two classes. It models the probabilities of one class. Logistic regression is used to find the best fitting model.
